```python
#선 그래프
import matplotlib.pyplot as plt
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

#선 그래프, x축 years값 , y축 gdp값
plt.plot(years, gdp, color='green' , marker = 'o', linestyle='solid')

#제목설정
plt.title("GDP per capita")

#y축에 레이블 설정
plt.ylabel("dollars")
#png 이미지로 저장
plt.savefig("gdp_per_capita.png", dpi=600)
plt.show()
```

```python
#실숩. 수학함수 그래프
x = [x for x in range(-10,10)]
y = [2*t for t in x]

plt.plot(x, y, marker='o')
plt.axis([-20, 20, -20 , 20])
plt.show()
```

```python
#실수. 난수 그래프
import numpy as np

x = [x for x in range(1000)]
y = np.random.rand(1000) * 6 - 3

plt.figure(figsize=(12,4))
plt.title("Random Numbers")
plt.plot(x,y,'bo-')
plt.axis([0,1000,-3.5,3.5])
plt.show()
```

```python
#포맷 스트링을 활용한 차트 장식
x = [x for x in range(-12,20)]
y1 = [2*t for t in x]
y2 = [t**2 + 5 for t in x]
y3 = [-t**2 - 5 for t in x]

#빨간색 점선, 녹색 실선과 세모기호, 파랑 별표와 점선
plt.plot(x,y1,'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30,30,-30,30])
plt.show()
```

```python
#범례(legend)
x = [x for x in range(20)]
y = [x**2 for x in range(20)]
z = [x**3 for x in range(20)]

plt.plot(x,x,label='linear')
plt.plot(x,y,label='quadratic')
plt.plot(x,z,label='qubic')

plt.xlabel('x label')
plt.ylabel('x label')
plt.legend()
plt.show()
```

```python
#실습. 사인 그래프
import math
x = []
y = []

for angle in range(360) :
  x.append(angle)
  y.append(math.sin(math.radians(angle)))

plt.plot(x,y)
plt.title("SINE WAVE")
plt.show()
```

```python
#실습. 사인, 코사인 그래프
x = []
y = []
z = []

for angle in range(360) :
  x.append(angle)
  z.append(math.cos(math.radians(angle)))
  y.append(math.sin(math.radians(angle)))

plt.plot(x,y)
plt.plot(x,z,'r-')
plt.title("SINE & COSINE WAVE")
plt.show()
```

```python
#정규분포 난수 히스토그램
mu1, sigma1 = 10, 2
mu2, sigma2 = -6, 3

standard_Gauss = np.random.randn(10000)
Gauss1 = mu1 + sigma1 * np.random.randn(10000)
Gauss2 = np.random.normal(mu2, sigma2, 10000)

plt.figure(figsize=(10,6))
plt.hist(standard_Gauss, bins = 50, alpha=0.4)    
plt.hist(Gauss1, bins = 50, alpha=0.4)
plt.hist(Gauss2, bins = 50, alpha=0.4)

plt.show()
```

```python
#균일분포, 정규분포, 표준정규분포
a = 2.0 * np.random.randn(10000) + 1.0  #정규분포 평1, 표편2 난수배열
b = np.random.standard_normal(10000)    #표준정규분포 평 0 표편1 난수배열
c = 20.0 * np.random.rand(10000) - 10.0#균일분포 -10.0 ~ 10.0 난수배열

plt.hist(a, bins=100, density = True, alpha = 0.7, histtype = 'step')
plt.hist(b, bins=50, density = True, alpha= 0.5, histtype = "stepfilled")
plt.hist(c, bins=100, density = True, alpha= 0.9, histtype = "step")

plt.show()
```

```python
#실습. 시그모이드 함수 그래프

x = np.linspace(-10,10,100)
y = 1/(1+np.exp(-x))

plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.show()
```

```python
#실습. 시그모이드 함수 미분값
def sigmoid(x) :
  s=1/(1+np.exp(-x))
  ds = s*(1-s)
  return s, ds
x = np.linspace(-10,10,100)
y1, y2 = sigmoid(x)

plt.figure(figsize=(8,4))
plt.plot(x,y1,x,y2)
plt.xlabel("x")
plt.ylabel("Sigmoid(x), Sigmoid'(x)")
plt.show()
```
